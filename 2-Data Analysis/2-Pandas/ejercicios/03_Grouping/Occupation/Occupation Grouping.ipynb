### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
ds = users.copy()
```

```python
ds = ds.groupby("gender").count()
ds
```

```python
def gender_to_numeric(x):
    if x == "M":
        return 1
    if x == "F":
        return 0
    
users["gender_n"] = users["gender"].apply(gender_to_numeric)

users
```

```python
def norm_by_data2(x):
    x['ratio'] = x["gender_n"].sum()/x["occupation"].count()
    return x

ds2 = users.groupby("occupation").apply(norm_by_data2)
ds2
```

```python
def norm_by_data2(x):
    x["ratio"] = x["user_id"]*100/x["user_id"].sum()
    return x

ds2 = users.reset_index().groupby(["occupation", "gender"]).count()
ds2.reset_index().groupby(['occupation']).apply(norm_by_data2)[['occupation', 'gender', 'ratio']]
```
